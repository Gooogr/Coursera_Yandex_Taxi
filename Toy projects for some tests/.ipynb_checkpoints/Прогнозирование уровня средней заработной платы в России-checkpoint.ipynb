**Визуальный анализ ряда**

```python
%pylab inline
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
plt.style.use('ggplot');
import warnings
from itertools import product
```

```python
def invboxcox(y, lmbda):
    '''
    Обратное преобразование Бокса - Кокса
    '''
    if lmbda == 0:
        return(np.exp(y))
    else:
        return(np.exp(np.log(lmbda*y+1)/lmbda))
```

```python
df = pd.read_csv('WAG_C_M.csv', sep = ';', index_col = 'month' ,parse_dates = True, dayfirst = True)
df.head()
```

```python
plt.figure(figsize(15,8))
df.WAG_C_M.plot()
plt.ylabel(u'Средняя номинальная заработная плата');
plt.xlabel(u'Год');
plt.title(u'Заработная плата в России в период  с января 1993 по август 2016');
```

Проверка стационарности и STL-декомпозиция ряда:

```python
plt.figure(figsize(15, 10))
sm.tsa.seasonal_decompose(df.WAG_C_M).plot();
```

```python
print 'Критерий Дики-Фуллера: ', sm.tsa.stattools.adfuller(df['WAG_C_M'])[1]
```

Наблюдаем устойчивый растущий тренд и годовую сезонность. 

**Стабилизация дисперсии**

Воспользуемся преобразованием Бокса-Кокса для стабилизации диспресии

```python
df['WAG_box'], lmbda = stats.boxcox(df.WAG_C_M)
plt.figure(figsize(15, 8))
df.WAG_box.plot()
```

```python
print 'Оптимальный параметр преобразования Бокса-Кокса: ', lmbda
print 'Критерий Дики-Фуллера: ', sm.tsa.stattools.adfuller(df['WAG_box'])[1]
```

**Выбор порядка дифференцирования**

Для начала используем сезонное дифференцирование. Исследуем результат про помощью STL-декомпозиции и критерия Дики-Фуллера

```python
df['WAG_diff12'] = df.WAG_box - df.WAG_box.shift(12)

plt.figure(figsize(15, 10))
sm.tsa.seasonal_decompose(df.WAG_diff12.dropna()).plot();
```

```python
print 'Критерий Дики-Фуллера: ', sm.tsa.stattools.adfuller(df['WAG_diff12'].dropna())[1]
```

Критерий Дики-Фуллера не отвергает гипотезу нестационарности, кроме того на графике тренда видно два сильных перегиба функции (кризисы 1998 года и 2008 - 2010 годов). Добавим к сезонному и обычное дифференцирование.

```python
df['WAG_diff12_1'] = df.WAG_diff12 - df.WAG_diff12.shift(1)

plt.figure(figsize(15, 10))
sm.tsa.seasonal_decompose(df.WAG_diff1.dropna()).plot();
```

```python
print 'Критерий Дики-Фуллера: ', sm.tsa.stattools.adfuller(df['WAG_diff12_1'].dropna())[1]
```

Гипотеза нестационарности уверенно отвергается, визуально тренд так же более не наблюдается.

**Выбор начальных приближений для p, q, P, Q**

Рассмотрим функции частичной и полной автокорреляции (ACF и PACF) для нашего стационарного ряда

```python
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(df['WAG_diff12_1'].dropna().values.squeeze(), lags = 48, ax = ax)
plt.show()
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(df['WAG_diff12_1'].dropna().values.squeeze(), lags = 48, ax = ax)
plt.show()
```

Начальные приближения:
* Q = 0 
* q = 5
* P = 1
* p = 5

**Обучение и сравнение моделей-кандидатов, выбор победителя**

```python
ps = range(0, 6)
d=1
qs = range(0, 6)
Ps = range(0, 2)
D=1
Qs = range(0, 2)
```

```python
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)
```

```python
%%time
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore') #отключим предупрежедения, так как не все модели, которые мы получим в процессе перебора, 
                                  #будут сходящимися
for param in parameters_list:
    #try except нужен, потому что на некоторых наборах параметров модель не обучается
    try:
        model=sm.tsa.statespace.SARIMAX(df.WAG_box, order=(param[0], d, param[1]), 
                                        seasonal_order=(param[2], D, param[3], 12)).fit(disp=-1)
    #выводим параметры, на которых модель не обучается и переходим к следующему набору
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    #сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
    
warnings.filterwarnings('default')
```

```python
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
```

```python
print(best_model.summary())
```

**Анализ остатков построенной модели**

Проверим модель на наличие автокоррелированности и смещенности

```python
plt.figure(figsize(15,8))
plt.subplot(211)
best_model.resid[13:].plot()
plt.ylabel(u'Остатки')

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(best_model.resid[13:].values.squeeze(), lags=48, ax=ax)

print("Критерий Стьюдента: p=%f" % stats.ttest_1samp(best_model.resid[13:], 0)[1])
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(best_model.resid[13:])[1])
```

Критерий Дики-Фуллера позволяет отклонить гипотезу о наличии автокоррелированности в остатках. Критерий Стьюдента не позволяет считать модель несмещенной. Скорректируем смещенность модели с помощью средненего значения остатков.

```python
mean_resid = np.abs(best_model.resid[13:].values.mean())
```
